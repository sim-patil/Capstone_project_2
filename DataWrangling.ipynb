```python
# Change all date time columns to datetime datatype
orders_v2["order_purchase_timestamp"] = pd.to_datetime(orders_v2["order_purchase_timestamp"], format='%Y-%m-%d %H:%M:%S')
orders_v2["order_aproved_at"] = pd.to_datetime(orders_v2["order_aproved_at"], format='%Y-%m-%d %H:%M:%S')
orders_v2["order_delivered_customer_date"] = pd.to_datetime(orders_v2["order_delivered_customer_date"], format='%Y-%m-%d %H:%M:%S')
orders_v2["order_estimated_delivery_date"] = pd.to_datetime(orders_v2["order_estimated_delivery_date"], format='%Y-%m-%d %H:%M:%S')
orders_v2["review_creation_date"] = pd.to_datetime(orders_v2["review_creation_date"], format='%Y-%m-%d %H:%M:%S')
orders_v2["review_answer_timestamp"] = pd.to_datetime(orders_v2["review_answer_timestamp"], format='%Y-%m-%d %H:%M:%S')
orders_v2.dtypes
```

```python
products_in_english.head()
```

```python
# Retrive the english name of products by mapping orders_v2 dataframe with product_category_name_translation dataframe
orders_v2['product_category_name'] = orders_v2['product_category_name'].map(products_in_english.set_index('product_category_name')['product_category_name_english']).fillna(orders_v2['product_category_name'])
```

```python
# delete '_' from the names and convert them to lowercase
orders_v2['product_category_name'] = orders_v2['product_category_name'].str.replace('_', ' ').str.lower()
orders_v2['product_category_name'].value_counts().head()
```

```python
# Calculate delivery delta
orders_v2['order_delivery_before_estimated_date_in_days'] = (orders_v2['order_estimated_delivery_date'] - orders_v2['order_delivered_customer_date']).dt.days

orders_v2.head()
```

```python
print("No. of orders delivered after estimated delivery date: ", sum(n < 0 for n in orders_v2.order_delivery_before_estimated_date_in_days))
print("No. of orders delivered before estimated delivery date: ",sum(n > 0 for n in orders_v2.order_delivery_before_estimated_date_in_days))
```

```python
# Calculate total order value by summing order value and freight value
orders_v2['total_order_value'] = orders_v2['order_products_value'].add(orders_v2['order_freight_value'])
orders_v2.head()
```
